```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def log_likelihood(X, y, weights):
    z = np.dot(X, weights)
    return np.sum(y * np.log(sigmoid(z)) + (1 - y) * np.log(1 - sigmoid(z)))

def gradient(X, y, weights):
    return np.dot(X.T, y - sigmoid(np.dot(X, weights)))

def train_logistic_regression(X, y, learning_rate, num_iterations):
    num_samples, num_features = X.shape
    weights = np.zeros(num_features)
    log_likelihoods = []
    percent_errors = []

    for i in range(num_iterations):
        y_pred = sigmoid(np.dot(X, weights))
        gradient_val = gradient(X, y, weights)
        weights += learning_rate * gradient_val

        log_likelihood_val = log_likelihood(X, y, weights)
        log_likelihoods.append(log_likelihood_val)

        predictions = (y_pred >= 0.5).astype(int)
        error_rate = np.mean(predictions != y)
        percent_errors.append(error_rate)

        print(f"Iteration {i+1}/{num_iterations}: Log-Likelihood = {log_likelihood_val}, Percent Error = {error_rate}")

    return weights, log_likelihoods, percent_errors

def load_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        data = [[int(bit) for bit in line.split()] for line in lines]
    return np.array(data)

train3_data = load_data('train3.txt')
train5_data = load_data('train5.txt')
```

```python
train3_labels = np.zeros(train3_data.shape[0])
train5_labels = np.ones(train5_data.shape[0])

X_train = np.vstack((train3_data, train5_data))
y_train = np.concatenate((train3_labels, train5_labels))
shuffle_indices = np.random.permutation(X_train.shape[0])
X_train = X_train[shuffle_indices]
y_train = y_train[shuffle_indices]

X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))

learning_rate = 0.001
num_iterations = 1000
weights, log_likelihoods, percent_errors = train_logistic_regression(X_train, y_train, learning_rate, num_iterations)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_iterations + 1), log_likelihoods)
plt.title('Log-Likelihood vs Iterations')
plt.xlabel('Iterations')
plt.ylabel('Log-Likelihood')

plt.subplot(1, 2, 2)
plt.plot(range(1, num_iterations + 1), percent_errors)
plt.title('Percent Error Rate vs Iterations')
plt.xlabel('Iterations')
plt.ylabel('Percent Error Rate')

plt.tight_layout()
plt.show()

learned_weights_matrix = weights[1:].reshape(8, 8)
print("\nLearned Weight Vector:")
print(learned_weights_matrix)

test3_data = load_data('test3.txt')
test5_data = load_data('test5.txt')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation function
def activate(z):
    return 1 / (1 + np.exp(-z))

# Compute the likelihood of the current model
def compute_log_likelihood(features, targets, coefficients):
    pred = features @ coefficients
    return np.sum(targets * np.log(activate(pred)) + (1 - targets) * np.log(1 - activate(pred)))

# Calculate the direction and magnitude of the update
def calculate_gradient(features, targets, coefficients):
    predictions = activate(features @ coefficients)
    return features.T @ (targets - predictions)

# Main routine for fitting the logistic model
def optimize_logistic_regression(features, targets, rate, iterations):
    sample_size, feature_count = features.shape
    coef = np.zeros(feature_count)
    likelihood_history = []
    error_history = []

    for iteration in range(iterations):
        pred = activate(features @ coef)
        coef += rate * calculate_gradient(features, targets, coef)

        likelihood = compute_log_likelihood(features, targets, coef)
        likelihood_history.append(likelihood)

        pred_labels = (pred >= 0.5).astype(int)
        error = np.mean(pred_labels != targets)
        error_history.append(error)

    return coef, likelihood_history, error_history

# Function to read the data from text files
def read_data(path):
    return np.loadtxt(path, dtype=int)

# Load and prepare training data
digit3_training = read_data('train3.txt')
digit5_training = read_data('train5.txt')

labels_digit3 = np.zeros(digit3_training.shape[0])
labels_digit5 = np.ones(digit5_training.shape[0])

features_train = np.vstack((digit3_training, digit5_training))
labels_train = np.concatenate((labels_digit3, labels_digit5))
ordering = np.random.permutation(features_train.shape[0])
features_train = features_train[ordering]
labels_train = labels_train[ordering]

features_train = np.insert(features_train, 0, 1, axis=1)

# Set hyperparameters and run the optimization
rate_of_learning = 0.001
num_of_iterations = 500
coefficients, likelihood_vals, error_rates = optimize_logistic_regression(features_train, labels_train, rate_of_learning, num_of_iterations)

# Visualization of the training process
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_of_iterations + 1), likelihood_vals)
plt.title('Log-Likelihood over Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Value of Log-Likelihood')

plt.subplot(1, 2, 2)
plt.plot(range(1, num_of_iterations + 1), error_rates)
plt.title('Error Rate over Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Error Rate')

plt.tight_layout()
plt.show()

# Display learned coefficients as a 2D matrix
coef_matrix = coefficients[1:].reshape(8, 8)
print("\nMatrix of Learned Coefficients:")
print(coef_matrix)

# Load test data and evaluate model performance
digit3_test = read_data('test3.txt')
digit5_test = read_data('test5.txt')

test3_labels = np.zeros(digit3_test.shape[0])
test5_labels = np.ones(digit5_test.shape[0])

# Stack the test data and labels
X_test = np.vstack((digit3_test, digit5_test))
y_test = np.concatenate((test3_labels, test5_labels))

# Add the bias term to the test data
X_test = np.insert(X_test, 0, 1, axis=1)

# Predict labels for test data and Calculate the error rate on the all test data
y_pred_test = (activate(X_test @ coefficients) >= 0.5).astype(int)

error_rate_test = np.mean(y_pred_test != y_test)
print(f"\nPercent Error Rate on All Testing Data: {error_rate_test}")

# Predict labels for test data and Calculate the error rate on the digit3 test data

y_pred_test3 = (activate(np.insert(digit3_test, 0, 1, axis=1) @ coefficients) >= 0.5).astype(int)

# Calculate the error rate on the test data
error_rate_test3 = np.mean(y_pred_test3 != test3_labels)

print(f"\nPercent Error Rate on digit 3 Testing Data: {error_rate_test3}")

# Predict labels for test data and Calculate the error rate on the digit5 test data

y_pred_test5 = (activate(np.insert(digit5_test, 0, 1, axis=1) @ coefficients) >= 0.5).astype(int)

# Calculate the error rate on the test data
error_rate_test5 = np.mean(y_pred_test5 != test5_labels)

print(f"\nPercent Error Rate on digit 5 Testing Data: {error_rate_test5}")
```
