```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
spectX_f = 'X.txt'
spectY_f = 'Y.txt'
X = np.loadtxt(spectX_f, dtype=int)
Y = np.loadtxt(spectY_f, dtype=int)
```

```python
# initialize pi to 0.05
p = np.ones((23,1)) * 0.05
```

```python
# compute the negative of the log likelihood
def c_loss(p,X,Y):
    temp = X.dot(np.log(1-p))
    return -1 / X.shape[0] * np.sum(Y[:,np.newaxis] * np.log(1-np.exp(temp)) + (1-Y[:,np.newaxis]) * temp)
```

```python
print(X.shape)
```

```python
# EM update for the parameter pi
def M_update(p,X,Y):
    temp0 = X.dot(np.log(1-p))
    temp1 = (X.T * p).T * Y[:,np.newaxis] / (1 - np.exp(temp0))
    p_new = (np.sum(temp1,axis = 0) / np.sum(X,axis = 0))[:,np.newaxis]
    return p_new
```

```python
# compute the number of mistakes made for every iteration
def predict(p,X,Y):
    temp = X.dot(np.log(1-p))
    y_prob = 1 - np.exp(temp)
    y_pred = np.where(y_prob >= 0.5,1,0)
    return np.sum(np.abs(y_pred - Y[:,np.newaxis]))
```

```python
# inital loss and initial prediction
initial_loss = c_loss(p,X,Y)
initial_error = predict(p,X,Y)
loss_list,error_list = [initial_loss],[initial_error]
```

```python
# learning steps
print("iteration 0\tnumber of mistiakes %d\tloss %.5f"%(initial_error,-initial_loss))
import math
max_iteration = 256
for i in range(1,max_iteration+1):
    p = M_update(p,X,Y)
    error = predict(p,X,Y)
    loss = c_loss(p,X,Y)
    error_list.append(error)
    loss_list.append(loss)
    if math.log(i, 2).is_integer():
        print("iteration %d\tnumber of mistiakes %d\tloss %.5f"%(i,error,-loss))
```

```python
error_list = np.array(error_list)
loss_list = np.array(loss_list)
x = np.linspace(0,256,257)
```

#### Since I compute the negative of the log likelihood, which is $loss = -\frac{1}{T}*\sum_{i=1}^{T}log\ P(Y=y^{{(t)}}|X=\vec {x^{(t)}})$, the plot of the loss will be symmetrical to the original log likelilhood with respect to x axis.

```python
fig=plt.figure(figsize=(30,10))
fig.add_subplot(1,2,1)
plt.plot(x,loss_list,'b')
plt.xlabel('number of iterations', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('loss', fontdict={'family' : 'Times New Roman', 'size'   : 20})
fig.add_subplot(1,2,2)
plt.plot(x,error_list,'g')
plt.xlabel('number of iterations', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('error rate', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```

```python
import numpy as np

# Load feature and target data from files
feature_data = np.loadtxt('X.txt')
target_data = np.loadtxt('Y.txt')
# Initial probability estimates
prob_estimates = np.full((23,1), 0.5)

# Function to calculate the likelihood of predictions
def calculate_log_likelihood(probabilities, features, targets):
    likelihood_product = np.prod((1 - probabilities.T) ** features, axis=1)
    log_likelihood_values = np.log(np.multiply(np.power(1 - likelihood_product,targets), np.power(likelihood_product, 1 - targets)))
    return np.mean(log_likelihood_values)

# Function to update probability estimates based on current predictions
def revise_estimates(probabilities, features, targets):
    product_denominator = 1 - np.prod((1 - probabilities.T) ** features, axis=1)
    temp_product = probabilities.T * features
    normalized_product = temp_product.T / product_denominator.reshape(-1,1).T
    final_product = normalized_product.T @ targets
    total_feature_sum = np.sum(features, axis=0)
    new_probabilities = (final_product / total_feature_sum).T
    return new_probabilities

# Function to count the number of incorrect predictions
def count_errors(probabilities, features, targets):
    predictions_product = 1 - np.prod((1 - probabilities.T) ** features, axis=1)
    predictions = np.where(predictions_product >= 0.5, 1, 0)
    error_count = np.sum(np.abs(predictions - targets))
    return error_count

# Display initial number of errors and loss
print(f"Iteration {0}\tErrors {count_errors(prob_estimates, feature_data, target_data)}\tLoss {calculate_log_likelihood(prob_estimates, feature_data, target_data):.5f}")

# Checkpoints for printing the performance metrics
for iteration in range(1,257):
    prob_estimates = revise_estimates(prob_estimates, feature_data, target_data)
    if (iteration in [1, 2, 4, 8, 16, 32, 64, 128, 256]):
        error_count = count_errors(prob_estimates, feature_data, target_data)
        loss_value = calculate_log_likelihood(prob_estimates, feature_data, target_data)
        print(f"Iteration {iteration}\tErrors {error_count}\tLoss {loss_value:.5f}")
```

```python
import numpy as np

X = np.loadtxt('X.txt')
Y = np.loadtxt('Y.txt')
P = np.ones((23,1)) * 0.05

def loglikelihood(P,X,Y):
    Pt = np.prod(np.power(1 - np.transpose(P), X), axis = 1)
    Pt = np.log(np.multiply(np.power(1 - Pt, Y), np.power(Pt, 1 - Y)))
    return np.mean(Pt)

def prob_update(P,X,Y):
    denom = 1 - np.prod(np.power(1 - np.transpose(P), X), axis = 1)
    Pt = np.matmul(np.transpose(np.transpose(np.divide(np.transpose(np.multiply(
        np.transpose(P), X)), np.transpose(denom.reshape(-1,1))))), Y)
    return np.transpose(Pt)/np.sum(X, axis = 0)

def mistakes(P,X,Y):
    pred = np.where(1 - np.prod(np.power(1 - np.transpose(P), X), axis = 1)>= 0.5, 1, 0)
    return np.sum(np.abs(pred - Y))

print(f"iteration 0\tmistakes {int(mistakes(P, X, Y))}\tloss {loglikelihood(P, X, Y):.5f}")

for t in range(1,257):
    P = prob_update(P,X,Y)
    if t in [1,2,4,8,16,32,64,128,256]:
        print(f"iteration {t}\tmistakes {int(mistakes(P, X, Y))}\tloss {loglikelihood(P, X, Y):.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.log(np.cosh(x))

# Define the derivative of f
def df(x):
    return np.tanh(x)

# Define the second derivative of f
def ddf(x):
    return 1 / np.cosh(x)**2

# Define the auxiliary function Q(x, y)
def Q(x, y):
    return f(y) + df(y) * (x - y) + 0.5 * (x - y)**2

# Define a range of x values for plotting
x_values = np.linspace(-10, 10, 400)

# Calculate the function values for f(x), Q(x, -2), Q(x, 3)
f_values = f(x_values)
Q_neg2_values = Q(x_values, -2)
Q_3_values = Q(x_values, 3)

# Plot f(x), Q(x, -2), and Q(x, 3)
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x)')
plt.plot(x_values, Q_neg2_values, label='Q(x, -2)')
plt.plot(x_values, Q_3_values, label='Q(x, 3)')
plt.legend()
plt.xlabel('x')
plt.ylabel('Function values')
plt.title('Plot of f(x) and Q(x, y) for y = -2 and y = 3')
plt.grid(True)
plt.show()
```

```python
# Define the update rule as a Python function
def update_xn(xn):
    return xn - np.tanh(xn)

# Initialize x0 for two different starting points
x0_neg2 = -2
x0_3 = 3

# Set a maximum number of iterations
max_iter = 20

# Create lists to store the sequence of x values for each initial guess
x_vals_neg2 = [x0_neg2]
x_vals_3 = [x0_3]

# Iterate and update x values using the update rule
for _ in range(max_iter):
    x0_neg2 = update_xn(x0_neg2)
    x0_3 = update_xn(x0_3)
    x_vals_neg2.append(x0_neg2)
    x_vals_3.append(x0_3)

# Plot the convergence for both starting points
plt.figure(figsize=(10, 6))
plt.plot(x_vals_neg2, label='x0 = -2')
plt.plot(x_vals_3, label='x0 = 3')
plt.xlabel('Iteration (n)')
plt.ylabel('x_n')
plt.title('Convergence of x_n over iterations')
plt.legend()
plt.grid(True)
plt.show()

# Output the last 5 values for each sequence to inspect convergence
x_vals_neg2[-5:], x_vals_3[-5:]
```

#### Using Newton Method, if x's initial value is too far away from the global minimum ($x_0 = -2$ or $x_0 = 3$), the update will not converge eventually.

```python
# x0 = -2
max_iteration = 20
x = -2
x3_list = [-2]
for i in range(max_iteration):
    x = x - 0.25*(np.exp(2*x)-np.exp(-2*x))
    x3_list.append(x)
x3_list = np.array(x3_list)
print(x3_list)
```

```python
# x0 = 3
max_iteration = 20
x = 1.09
x4_list = [1.09]
for i in range(max_iteration):
    x = x - 0.25*(np.exp(2*x)-np.exp(-2*x))
    x4_list.append(x)
x4_list = np.array(x4_list)
print(x4_list)
```

#### In order to converge we want the find the upper bound of $|x_0|$ so that after the first update, $|x_1| <= |x_0|$. 

```python
x = np.linspace(-5,5,10000)
y = x - 0.25 * (np.exp(2*x)-np.exp(-2*x))
z = np.abs(x)-np.abs(y)
index = np.where(z>0)
bound = np.abs(x[index][0])
print("upper bound for |x0| is %.4f"%bound)
```

#### Plots for $g(x) = \frac{1}{10}\sum_{k=1}^{10}log\ cosh(x+\frac{2}{\sqrt{k}})$.

```python
x = np.linspace(-100,100,1000)
y = np.zeros(1000)
for i in range(1,11):
    y = y + np.log(np.cosh(x+2/(i**0.5)))
y = y / 10
plt.plot(x,y,'b')
plt.xlabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('g(x)', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```

#### Update rule $x_{n+1} = x_n - \frac{1}{10}\sum_{k=1}^{10}{sinh(x_n+\frac{2}{\sqrt{i}})*cosh(x_n+\frac{2}{\sqrt{i}})^{-1}}$

```python
max_iteration = 50
x = -10
x5_list = [-10]
for i in range(max_iteration):
    gradient = 0
    for j in range(1,11):
        gradient += np.sinh(x+2/j**0.5)/np.cosh(x+2/j**0.5)
    x = x - 1/10 * gradient
    x5_list.append(x)
x5_list = np.array(x5_list)
result = 0
for i in range(1,11):
    result = result + np.log(np.cosh(x5_list[-1]+2/(i**0.5)))
result = result / 10
print("the minimum of g(x) = %.4f when x = %.4f"%(result,x5_list[-1]))
```
