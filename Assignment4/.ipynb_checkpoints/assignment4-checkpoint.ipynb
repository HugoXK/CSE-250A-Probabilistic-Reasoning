```python
# Question 4.3 (a)

"""
compute the maximum likelihood estimate of the unigram distribution Pu(w) over words w. Print out a table 
of all the tokens that start with the letter “M”, along with their numerical unigram probabilities
"""

# Define the file paths (assuming they're in the same directory as the script)
vocab_file_path = 'hw4_vocab.txt'
unigram_file_path = 'hw4_unigram.txt'

# Read the vocabulary into a list
with open(vocab_file_path, 'r') as vocab_file:
    vocab_list = [line.strip() for line in vocab_file]

# Initialize a variable for the total count of all unigrams
total_count = 0

# Initialize a dictionary to hold the unigram counts indexed by the line number
unigram_counts = {}

# Read the unigram counts file and populate the dictionary
# Also calculate the total counts on the fly
with open(unigram_file_path, 'r') as unigram_file:
    for i, line in enumerate(unigram_file):
        count = int(line.strip())
        unigram_counts[i+1] = count  # +1 because the file line index starts at 1
        total_count += count

# Now, filter the vocab_list for words that start with 'M' and collect their counts
m_word_counts = {word: unigram_counts[vocab_list.index(word) + 1] for word in vocab_list if word.startswith('M')}

# Calculate and print out the MLE probabilities for the 'M' words
print("TOKEN\t\tPROBABILITY")
for word in sorted(m_word_counts.keys()):  # Sorted to print in alphabetical order
    probability = m_word_counts[word] / total_count
    print(f"{word}\t\t{probability:.6f}")
```

```python
# Question 4.3 (b)

"""
compute the maximum likelihood estimate of the bigram distribution Pb(w′|w). Print out a table of
the ten most likely words to follow the word “THE”, along with their numerical bigram probabilities.
"""

# Define the file paths
vocab_file_path = 'hw4_vocab.txt'
unigram_file_path = 'hw4_unigram.txt'
bigram_file_path = 'hw4_bigram.txt'

# Read the vocabulary into a list
with open(vocab_file_path, 'r') as vocab_file:
    vocab_list = [line.strip() for line in vocab_file]

# Get the index of "THE" in the vocabulary list
the_index = vocab_list.index('THE') + 1  # Adding 1 to match line numbering

# Initialize a dictionary to hold the unigram counts
unigram_counts = {}

# Read the unigram counts file and populate the dictionary
with open(unigram_file_path, 'r') as unigram_file:
    for i, line in enumerate(unigram_file):
        unigram_counts[i+1] = int(line.strip())  # Dictionary with line number as key

# Read the bigram counts and filter for those that start with "THE"
the_bigram_counts = {}
with open(bigram_file_path, 'r') as bigram_file:
    for line in bigram_file:
        parts = line.strip().split()
        if int(parts[0]) == the_index:  # Check if the first word is "THE"
            w_index = int(parts[1])
            count = int(parts[2])
            the_bigram_counts[vocab_list[w_index - 1]] = count  # Store with actual word as key

# Now calculate the probabilities
the_bigram_probabilities = {}
for word, count in the_bigram_counts.items():
    the_bigram_probabilities[word] = count / unigram_counts[the_index]

# Get the top 10 words most likely to follow "THE" by their bigram probability
top_the_bigrams = sorted(the_bigram_probabilities.items(), key=lambda item: item[1], reverse=True)[:10]

# Print out the table
print("TOKEN\t\tPROBABILITY")
for word, prob in top_the_bigrams:
    print(f"{word}\t\t{prob:.6f}")
```

```python
# Question 4.3 (c)

"""
ignore punctuation in the sentence “The stock market fell by one hundred points last week.”,
compute and compare the log-likelihoods of this sentence under the unigram and bigram models.
"""

import math

# Load the vocabulary
with open('hw4_vocab.txt', 'r') as f:
    vocab = f.read().splitlines()

# Load the unigram counts
with open('hw4_unigram.txt', 'r') as f:
    unigram_counts = [int(count.strip()) for count in f.readlines()]

# Load the bigram counts into a dictionary for quick access
bigram_counts = {}
with open('hw4_bigram.txt', 'r') as f:
    for line in f:
        w1, w2, count = [int(x) for x in line.strip().split()]
        bigram_counts[(w1, w2)] = count

# Calculate total number of words
total_word_count = sum(unigram_counts)

# Function to calculate the probability of a unigram
def unigram_probability(word):
    index = vocab.index(word) if word in vocab else -1
    return unigram_counts[index] / total_word_count if index >= 0 else 1e-10

# Function to calculate the probability of a bigram
def bigram_probability(w1, w2):
    w1_index = vocab.index(w1) if w1 in vocab else -1
    w2_index = vocab.index(w2) if w2 in vocab else -1
    bigram_count = bigram_counts.get((w1_index+1, w2_index+1), 0)
    w1_count = unigram_counts[w1_index] if w1_index >= 0 else 1e-10
    return bigram_count / w1_count if w1_count > 0 else 1e-10

# Define the sentence
sentence = "The stock market fell by one hundred points last week"

# Tokenize and convert to uppercase (assuming vocabulary is in uppercase)
tokens = ['<s>'] + sentence.upper().split()

# Calculate the log-likelihoods for unigram and bigram models
log_likelihood_unigram = sum(math.log(unigram_probability(token)) for token in tokens[1:])
log_likelihood_bigram = sum(math.log(bigram_probability(w1, w2)) for w1, w2 in zip(tokens[:-1], tokens[1:]))

# Print the log-likelihoods
print(f"Unigram log-likelihood (Lu): {log_likelihood_unigram}")
print(f"Bigram log-likelihood (Lb): {log_likelihood_bigram}")

# Compare log-likelihoods
if log_likelihood_unigram > log_likelihood_bigram:
    print("The unigram model yields the highest log-likelihood.")
else:
    print("The bigram model yields the highest log-likelihood.")
```

```python
# Question 4.3 (d)

"""
ignore punctuation, compute and compare the log-likelihoods of this sentence under the unigram and bigram models
pairs of adjacent words in this sentence are not observed in the training corpus, effect on the log-likelihood from the bigram model
"""

import numpy as np
import math

import numpy as np
import matplotlib.pyplot as plt
import math

# Load the vocabulary
with open('hw4_vocab.txt', 'r') as f:
    vocab = f.read().splitlines()

# Load the unigram counts
with open('hw4_unigram.txt', 'r') as f:
    unigram_counts = [int(count.strip()) for count in f.readlines()]

# Load the bigram counts into a dictionary for quick access
bigram_counts = {}
with open('hw4_bigram.txt', 'r') as f:
    for line in f:
        w1, w2, count = [int(x) for x in line.strip().split()]
        bigram_counts[(w1, w2)] = count

# Calculate total number of words
total_word_count = sum(unigram_counts)

# Convert the sentence into indices based on the provided vocabulary
sentence = "The sixteen officials sold fire insurance"
tokens = ['<s>'] + sentence.upper().split()
index_sentence = [vocab.index(token) if token in vocab else -1 for token in tokens]


# Function to calculate log-likelihood of a sentence using unigram model
def unigram_log_likelihood(sentence_tokens):
    log_likelihood = 0
    for token in sentence_tokens:
        prob = unigram_probability(token)
        log_likelihood += math.log(prob)
    return log_likelihood

def bigram_log_likelihood(sentence_tokens):
    log_likelihood = 0
    unobserved_pairs = []  # List to keep track of unobserved word pairs
    for i in range(len(sentence_tokens)-1):
        w1 = sentence_tokens[i]
        w2 = sentence_tokens[i+1]
        prob = bigram_probability(w1, w2)
        if prob > 0:
            log_likelihood += math.log(prob)
        else:
            log_likelihood += float('-inf')  # Log of zero probability for unobserved pairs
            unobserved_pairs.append((w1, w2))  # Add the unobserved pair to the list
    return log_likelihood, unobserved_pairs

# Compute log-likelihoods and unobserved pairs
unigram_ll = unigram_log_likelihood(tokens[1:])  # Exclude the <s> token for unigram model
bigram_ll, unobserved_pairs = bigram_log_likelihood(tokens)  # Include the <s> token for bigram model

# Output the log-likelihoods and unobserved pairs
print(f"Unigram log-likelihood (Lu): {unigram_ll}")
print(f"Bigram log-likelihood (Lb): {bigram_ll}")
if unobserved_pairs:
    print("\nThe bigrams which were not observed in the training corpus:")
    for pair in unobserved_pairs:
        print(' '.join(pair))
    print("made the product of all the probabilities 0, which made the log-likelihood -infinity.")
else:
    print("All bigram pairs were observed in the training corpus.")
```

```python
# Question 4.3 (e)

"""
mixture model that predicts words from a weighted interpolation of the unigram and bigram models
compute and plot the value of this log-likelihood Lm as a function of the parameter λ ∈ [0, 1].
"""

import numpy as np
import matplotlib.pyplot as plt
import math

# Load the vocabulary
with open('hw4_vocab.txt', 'r') as f:
    vocab = f.read().splitlines()

# Load the unigram counts
with open('hw4_unigram.txt', 'r') as f:
    unigram_counts = [int(count.strip()) for count in f.readlines()]

# Load the bigram counts into a dictionary for quick access
bigram_counts = {}
with open('hw4_bigram.txt', 'r') as f:
    for line in f:
        w1, w2, count = [int(x) for x in line.strip().split()]
        bigram_counts[(w1, w2)] = count

# Calculate total number of words
total_word_count = sum(unigram_counts)

# Convert the sentence into indices based on the provided vocabulary
sentence = "The sixteen officials sold fire insurance"
tokens = ['<s>'] + sentence.upper().split()
index_sentence = [vocab.index(token) if token in vocab else -1 for token in tokens]

# Define the function to calculate the interpolated probability
def Pm(w1_index, w2_index, lambda_value):
    uni_prob = unigram_probability(vocab[w2_index]) if w2_index >= 0 else 1e-10
    bi_prob = bigram_probability(vocab[w1_index], vocab[w2_index]) if w1_index >= 0 else 1e-10
    return lambda_value * uni_prob + (1 - lambda_value) * bi_prob

# Define the unigram_probability and bigram_probability functions based on the user input
def unigram_probability(word):
    index = vocab.index(word) if word in vocab else -1
    return unigram_counts[index] / total_word_count if index >= 0 else 1e-10

def bigram_probability(w1, w2):
    w1_index = vocab.index(w1) if w1 in vocab else -1
    w2_index = vocab.index(w2) if w2 in vocab else -1
    bigram_count = bigram_counts.get((w1_index+1, w2_index+1), 0)
    w1_count = unigram_counts[w1_index] if w1_index >= 0 else 1e-10
    return bigram_count / w1_count if w1_count > 0 else 1e-10

# Calculate the total number of words
total_word_count = sum(unigram_counts)

# Lambda values to test
lambda_list = np.linspace(0.00001, 0.99999, 1000)
log_likelihoods = []

# Calculate the log likelihood for each lambda
for lambda_value in lambda_list:
    log_likelihood = 0
    for i in range(1, len(index_sentence)):
        prob = Pm(index_sentence[i-1], index_sentence[i], lambda_value)
        log_likelihood += np.log(prob)
    log_likelihoods.append(log_likelihood)

# Find the lambda that gives the highest log likelihood
best_lambda = lambda_list[np.argmax(log_likelihoods)]
best_log_likelihood = max(log_likelihoods)

# Plotting
plt.plot(lambda_list, log_likelihoods, 'b')
plt.xlabel('Lambda', fontdict={'family': 'Times New Roman', 'size': 12})
plt.ylabel('Log Likelihood', fontdict={'family': 'Times New Roman', 'size': 12})
plt.title('Lambda vs Log Likelihood')
plt.show()

print(f"Optimal lambda: {best_lambda:.2f}")
print(f"Maximum log-likelihood with the mixture model: {best_log_likelihood:.2f}")
```

```python
# Question 4.4 (a)

"""
using only data from the year 2000, compute the linear coefficients (a1,a2,a3) that maximize the log-likelihood
"""

import numpy as np
from scipy.optimize import minimize

# Function to load data from a text file
def load_data(file_name):
    return np.loadtxt(file_name)

# Function to compute the log-likelihood
def log_likelihood(coeffs, data):
    a1, a2, a3 = coeffs
    likelihood = 0
    for t in range(3, len(data)):
        xt = data[t]
        xt_1 = data[t-1]
        xt_2 = data[t-2]
        xt_3 = data[t-3]
        predicted_xt = a1*xt_1 + a2*xt_2 + a3*xt_3
        likelihood += -0.5 * (xt - predicted_xt)**2
    return -likelihood

# Load the data from text files
nasdaq00_data = load_data('nasdaq00.txt')
nasdaq01_data = load_data('nasdaq01.txt')

# Initial guess for the coefficients
initial_guess = [0.1, 0.1, 0.1]

# Minimize the negative log-likelihood
result = minimize(log_likelihood, initial_guess, args=(nasdaq00_data,), method='L-BFGS-B')

# The optimal coefficients are
coefficients = result.x
print("The optimal coefficients are:", coefficients)
```
